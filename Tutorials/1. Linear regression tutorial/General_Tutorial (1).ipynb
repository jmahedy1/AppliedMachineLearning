# **Introduction on CoLab**
Google Colaboratory (Colab) can be considered a free CPU/ GPU/ TPU-supported Jupyter notebook environment that runs entirely in the cloud. It does not require a setup and supports many popular machine learning libraries which can be easily loaded in your notebook.
1. How to start? Answer: Set up a Google account if you don't already have one, (1) login to your Google account, and go to your Google Drive, (2) click "New"->"More"->"Google Colaboratory". An alternative way: (1) Login to your Google Drive, (2) Open the following URL in your browser: https://colab.research.google.com, (3) Click on the NEW PYTHON 3 NOTEBOOK link at the bottom of the screen.
2. How to leverage cloud GPU? Answer: On Menu, click "Edit"->"Notebook Settings" -> "Hardware Accelerator" -> "GPU".
3. How to run all the cells before the current cell? Answer: On Menu, click "Runtime"->"Run before". 
4. How to run the entire code in your notebook without an interruption? Answer: On Menu, click "Runtime"->"Restart and Run all"
5. Any usage requirement? Answer: Ensure to use Colab with a stable internet since it runs on cloud based servers. 
6. How to save the codes running in Colab? Answer: All the codes are automatically saved on your Google Drive if you are connected to the Internet. You can also click "File"-> "Save a copy in Drive".
7. Colab allows to have equations in the comment by using Markdown that is a lightweight markup language with plain-text-formatting syntax. For example, $\sqrt{3x-1}+(1+x)^2$

# **Useful Python Libraries for ML**

2. **Scikit-learn**: A machine learning library for Python programming language. It supports a range of classification, regression, and clustering algorithms. 
3. **Pandas**: A software library written for the Python programming language for data manipulation and analysis. In particular, it offers data structures and operations for manipulating numerical tables and time series.
4. **Matplotlib**: A plotting library for the Python programming language and its numerical mathematics extension NumPy.
